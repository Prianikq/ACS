# Задание 1 - Модель "Хищник-Жертва"

```python
y0 = [2, 1]
plt = plot(0)
var("y1, y2")
colors = (x for x in ["red", "green", "blue"])
for b in range(4,1,-1):
    a, c, d = 3, 1, 1
    f = [y1*(a-b*y2),y2*(-c+d*y1)]
    sol = desolve_odeint(f, y0, times = srange(0, 7, 0.01), dvars = [y1, y2])
    plt += line(sol, color = next(colors))
show(plt)  
```

# Задание 2 - Падение тела, брошенного под углом к горизонту

```python
tmax = 1.41 # время движения, подбирается экспериментально
alph = pi/4 # угол бросания тела
v0 = 10.0 # начальная скорость
y0 = [0, 0, v0*cos(alph), 0, v0*sin(alph) ] # начальные условия
var("y1, y2, y3, y4, t")
k = 0.01
g = 9.81
f = [y2, -k*y2*sqrt(y2**2+y4**2), y4, -k*y4*sqrt(y2**2+y4**2)-g]
sol = desolve_system_rk4(des = f, vars = [y1, y2, y3, y4], ics = y0, ivar = t, step = 0.01, end_points = tmax)
plt = plot(0)
for i in range(len(sol) - 1):
    plt += line([[sol[i][1], sol[i][3]], [sol[i + 1][1], sol[i + 1][3]]])
show(plt)
```
